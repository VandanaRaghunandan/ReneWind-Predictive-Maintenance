The target variable represents:

- **0** → No Generator Failure

In real-world predictive maintenance datasets, failure events are usually much rarer than normal operating conditions. This results in an **imbalanced dataset**, where one class significantly outnumbers the other.

### Why is Class Imbalance a Problem?

For the ReneWind project, class imbalance is particularly important because the business costs are highly unequal.

| Prediction Outcome | Business Cost |
|--------------------|--------------:|
| False Negative (Missed Failure) | **$200,000 Loss** |
| False Positive (False Alarm) | **$5,000 Cost** |
| True Positive (Correct Failure Prediction) | **$20,000 Benefit** |

A model that simply predicts every turbine as **"No Failure"** may achieve a high accuracy because failures are rare, but it would miss most actual failures. Each missed failure can lead to catastrophic financial losses.

Therefore, throughout this project, we will focus on identifying failures effectively rather than relying solely on overall accuracy.

```python
# ==========================================================
# Exploratory Data Analysis (EDA)
# Class Distribution Analysis
# ==========================================================

# ----------------------------------------------------------
# Step 1 : Calculate Class Distribution (Raw Counts)
# ----------------------------------------------------------

# Count the number of observations in each target class
class_counts = train_df['Target'].value_counts()

# Display raw counts
print("=" * 50)
print("Class Distribution (Count)")
print("=" * 50)
print(class_counts)

# ----------------------------------------------------------
# Step 2 : Calculate Class Distribution (Percentage)
# ----------------------------------------------------------

# Convert counts into percentages
class_percentages = train_df['Target'].value_counts(normalize=True) * 100

# Display percentages
print("\n" + "=" * 50)
print("Class Distribution (Percentage)")
print("=" * 50)
print(class_percentages.round(2))

# ----------------------------------------------------------
# Step 3 : Visualize Class Distribution
# ----------------------------------------------------------

# Create figure
plt.figure(figsize=(7,5))

# Draw countplot
ax = sns.countplot(
    x='Target',
    data=train_df,
    palette='Set2'
)

# Set title
plt.title("Class Distribution of Generator Failure", fontsize=14)

# Set axis labels
plt.xlabel("Target Class")
plt.ylabel("Number of Samples")

# Rename x-axis labels
ax.set_xticklabels(["No Failure (0)", "Failure (1)"])

# ----------------------------------------------------------
# Step 4 : Display Exact Count on Top of Each Bar
# ----------------------------------------------------------

# Loop through each bar
for bar in ax.patches:

    # Get bar height
    height = bar.get_height()

    # Display value above bar
    ax.annotate(
        f'{int(height)}',
        (bar.get_x() + bar.get_width()/2, height),
        ha='center',
        va='bottom',
        fontsize=11,
        fontweight='bold'
    )

# Improve layout
plt.tight_layout()

# Display plot
plt.show()
```

# Exploratory Data Analysis (EDA) - Statistical Feature Comparison

## Objective

Visual inspection of feature distributions is useful, but statistical tests provide quantitative evidence of whether a feature differs significantly between the two target classes.

In this section, two statistical tests are performed for every feature.

### 1. Kolmogorov–Smirnov (KS) Test

The KS test compares the entire distributions of two groups.

For each feature, it measures the maximum difference between the cumulative distributions of:

- **No Failure (Target = 0)**
- **Failure (Target = 1)**

A **larger KS statistic** indicates that the feature distributions are more different, making the feature potentially more informative.

---

### 2. Welch's t-test

The t-test compares the mean values of the two groups.

Using Welch's version of the t-test allows for unequal variances between classes, which is more appropriate for real-world datasets.

A **small p-value** (typically < 0.05) suggests that the difference in means is statistically significant.

---

## Why are we doing this?

Instead of relying only on visual inspection, these statistical tests help us objectively identify which features are most useful for distinguishing between generator failures and normal operating conditions.

Features with:

- High KS statistic
- Small KS p-value
- Large absolute t-statistic
- Small t-test p-value

are strong candidates for predictive modeling.

```python
# ==========================================================
# Exploratory Data Analysis (EDA)
# Statistical Feature Ranking
# ==========================================================

# ----------------------------------------------------------
# Step 1 : Import Required Statistical Functions
# ----------------------------------------------------------

from scipy.stats import ks_2samp
from scipy.stats import ttest_ind

# ----------------------------------------------------------
# Step 2 : Create List of Features
# ----------------------------------------------------------

# Exclude the target column
feature_columns = [col for col in train_df.columns if col != "Target"]

# Create an empty list to store results
results = []

# ----------------------------------------------------------
# Step 3 : Perform Statistical Tests
# ----------------------------------------------------------

# Loop through every feature
for feature in feature_columns:

    # Separate feature values based on target class
    no_failure = train_df[train_df["Target"] == 0][feature]
    failure = train_df[train_df["Target"] == 1][feature]

    # -----------------------------
    # Kolmogorov-Smirnov Test
    # -----------------------------
    ks_statistic, ks_pvalue = ks_2samp(
        no_failure,
        failure
    )

    # -----------------------------
    # Welch's t-test
    # -----------------------------
    t_statistic, t_pvalue = ttest_ind(
        no_failure,
        failure,
        equal_var=False
    )

    # Store results
    results.append({
        "Feature": feature,
        "KS_statistic": ks_statistic,
        "KS_pvalue": ks_pvalue,
        "T_statistic": t_statistic,
        "T_pvalue": t_pvalue
    })
# ----------------------------------------------------------
# Step 4 : Create Results DataFrame
# ----------------------------------------------------------

# Convert results list into a DataFrame
results_df = pd.DataFrame(results)

# ----------------------------------------------------------
# Step 5 : Sort Features by KS Statistic
# ----------------------------------------------------------

# Sort features based on KS Statistic (highest first)
results_df = results_df.sort_values(
    by="KS_statistic",
    ascending=False
).reset_index(drop=True)

# ----------------------------------------------------------
# Step 6 : Assign Feature Rank
# ----------------------------------------------------------

# Add ranking column
results_df.insert(
    0,
    "Rank",
    range(1, len(results_df) + 1)
)

# ----------------------------------------------------------
# Step 7 : Display Top 15 Features
# ----------------------------------------------------------

print("=" * 95)
print("Top 15 Most Informative Features")
print("=" * 95)

display(results_df.head(15))

# ----------------------------------------------------------
# Step 8 : Plot Top 15 Features by KS Statistic
# ----------------------------------------------------------

# Select top 15 features
top15 = results_df.head(15)

# Create figure
plt.figure(figsize=(12, 7))

# Plot horizontal bar chart
sns.barplot(
    data=top15,
    x="KS_statistic",
    y="Feature",
    palette="viridis"
)

# Add chart title
plt.title(
    "Top 15 Features Ranked by KS Statistic",
    fontsize=15
)

# Label axes
plt.xlabel("KS Statistic")
plt.ylabel("Feature")

# Add KS values beside each bar
for index, value in enumerate(top15["KS_statistic"]):
    plt.text(
        value + 0.003,
        index,
        f"{value:.3f}",
        va="center",
        fontsize=10
    )

# Improve layout
plt.tight_layout()

# Display chart
plt.show()

# ----------------------------------------------------------
# Step 9 : Display Complete Ranking
# ----------------------------------------------------------

print("\nComplete Feature Ranking")

display(results_df)
```

# Interpretation

The statistical tests provide an objective ranking of feature importance based on how differently each feature behaves between the failure and non-failure classes.

### KS Statistic

The KS statistic ranges between **0 and 1**.

- Values close to **0** indicate that the two class distributions are very similar.
- Values closer to **1** indicate strong separation between the two classes.

Higher KS statistics suggest that a feature is more useful for distinguishing generator failures.

---

### KS p-value

The KS p-value indicates whether the observed distribution difference is statistically significant.

- **p < 0.05** → Significant difference
- **p ≥ 0.05** → No significant difference

---

### t-statistic

The t-statistic measures how different the mean values are between the two classes.

Large positive or negative values indicate greater separation between the class means.

---

### t-test p-value

The t-test p-value determines whether the observed difference in means is statistically significant.

A p-value less than **0.05** suggests that the feature has significantly different mean values for failure and non-failure observations.

---

## Conclusion

Features appearing near the top of the ranking exhibit the strongest statistical differences between the two classes.

These variables are expected to be among the most informative predictors for the machine learning models.

Later in the project, we will compare these statistically important features with feature importance obtained from machine learning algorithms such as Random Forest to validate their predictive value.

# Exploratory Data Analysis (EDA) - Correlation Analysis

## Objective

In this section, we analyze the relationships between the features in the ReneWind dataset.

Correlation measures the strength and direction of a linear relationship between two variables.

The Pearson correlation coefficient ranges from **-1 to +1**.

- **+1** → Perfect positive correlation
- **-1** → Perfect negative correlation
- **0** → No linear relationship

---

## Why is this important?

### 1. Feature-to-Feature Correlation

Highly correlated features contain similar information.

When two features are strongly correlated (for example, correlation > 0.80), they may introduce **multicollinearity**.

Multicollinearity can:

- Make model interpretation more difficult.
- Increase instability in linear models.
- Add redundant information.

Tree-based models such as Random Forest are generally less affected, but identifying highly correlated features remains good practice.

---

### 2. Feature-to-Target Correlation

The correlation between each feature and the target provides an initial indication of predictive power.

Features with larger absolute correlations are often more informative for classification, although correlation alone does not guarantee importance because machine learning models can also capture non-linear relationships.

This analysis complements the statistical tests performed in the previous section.

```python
# ==========================================================
# Exploratory Data Analysis (EDA)
# Correlation Analysis
# ==========================================================

# ----------------------------------------------------------
# Step 1 : Compute Correlation Matrix
# ----------------------------------------------------------

# Calculate Pearson correlation matrix
correlation_matrix = train_df.corr()

# ----------------------------------------------------------
# Step 2 : Plot Full Correlation Heatmap
# ----------------------------------------------------------

# Create figure
plt.figure(figsize=(18, 16))

# Draw heatmap
sns.heatmap(
    correlation_matrix,
    cmap="coolwarm",
    center=0,
    square=True,
    linewidths=0.5,
    cbar=True
)

# Add title
plt.title(
    "Correlation Heatmap of ReneWind Features",
    fontsize=18
)

# Display plot
plt.show()

# ----------------------------------------------------------
# Step 3 : Identify Highly Correlated Feature Pairs
# ----------------------------------------------------------

print("=" * 70)
print("Highly Correlated Feature Pairs (Correlation > 0.80)")
print("=" * 70)

# Create empty list
high_corr_pairs = []

# Remove target column
feature_corr = train_df.drop(columns=["Target"]).corr()

# Loop through upper triangle only
for i in range(len(feature_corr.columns)):
    for j in range(i + 1, len(feature_corr.columns)):

        corr_value = feature_corr.iloc[i, j]

        if abs(corr_value) > 0.80:

            high_corr_pairs.append([
                feature_corr.columns[i],
                feature_corr.columns[j],
                corr_value
            ])

# Print results
if len(high_corr_pairs) == 0:
    print("No feature pairs have correlation greater than 0.80.")
else:

    high_corr_df = pd.DataFrame(
        high_corr_pairs,
        columns=[
            "Feature 1",
            "Feature 2",
            "Correlation"
        ]
    )

    high_corr_df = high_corr_df.sort_values(
        by="Correlation",
        ascending=False
    )

    display(high_corr_df)

# ----------------------------------------------------------
# Step 4 : Compute Correlation with Target
# ----------------------------------------------------------

target_corr = (
    correlation_matrix["Target"]
    .drop("Target")
    .sort_values(
        key=abs,
        ascending=False
    )
)

# ----------------------------------------------------------
# Step 5 : Display Top Correlated Features
# ----------------------------------------------------------

print("=" * 70)
print("Top 10 Features Most Correlated with Target")
print("=" * 70)

display(target_corr.head(10))

# ----------------------------------------------------------
# Step 6 : Plot Top 10 Features
# ----------------------------------------------------------

plt.figure(figsize=(10, 6))

sns.barplot(
    x=target_corr.head(10).values,
    y=target_corr.head(10).index,
    palette="viridis"
)

plt.title(
    "Top 10 Features Most Correlated with Target",
    fontsize=15
)

plt.xlabel("Correlation with Target")
plt.ylabel("Feature")

# Add values beside bars
for index, value in enumerate(target_corr.head(10).values):

    plt.text(
        value,
        index,
        f"{value:.3f}",
        va="center",
        fontsize=10
    )

plt.tight_layout()

plt.show()
```

# Exploratory Data Analysis (EDA) - Outlier Analysis

## Objective

In this section, we investigate the presence of outliers in each sensor feature.

An outlier is an observation that lies far away from the majority of the data.

We will:

1. Visualize each feature using boxplots.
2. Detect outliers using the Interquartile Range (IQR) method.
3. Count the number of outliers in every feature.
4. Calculate the percentage of outliers.

---

## Why are outliers important in predictive maintenance?

In industrial sensor datasets, extreme values are not necessarily errors.

Outliers may represent:

- Sudden equipment degradation
- Mechanical wear
- Abnormal vibration
- Temperature spikes
- Electrical instability
- Early warning signs before generator failure

Some outliers may also arise due to:

- Sensor malfunction
- Transmission errors
- Measurement noise

Therefore, it is important to distinguish between genuine anomalies and data quality issues.

---

## Our Approach

For the ReneWind project, **we will not remove outliers**.

Since our objective is to predict generator failures, extreme sensor readings may contain valuable information that helps identify impending failures.

Tree-based machine learning algorithms such as Random Forest are generally robust to outliers, making it appropriate to retain these observations.

```python
# ==========================================================
# Exploratory Data Analysis (EDA)
# Outlier Detection using IQR Method
# ==========================================================

# ----------------------------------------------------------
# Step 1 : Create Feature List
# ----------------------------------------------------------

# Select all feature columns except the target
feature_columns = [col for col in train_df.columns if col != "Target"]

# ----------------------------------------------------------
# Step 2 : Plot Boxplots for All Features
# ----------------------------------------------------------

# Create figure with 8 rows and 5 columns
fig, axes = plt.subplots(
    nrows=8,
    ncols=5,
    figsize=(22, 30)
)

# Flatten axes
axes = axes.flatten()

# Loop through every feature
for i, feature in enumerate(feature_columns):

    # Draw boxplot
    sns.boxplot(
        y=train_df[feature],
        ax=axes[i],
        color="skyblue"
    )

    # Add subplot title
    axes[i].set_title(feature, fontsize=11)

    # Remove x-axis label
    axes[i].set_xlabel("")

    # Add y-axis label
    axes[i].set_ylabel("Value")

# Improve layout
plt.tight_layout()

# Display plots
plt.show()

# ----------------------------------------------------------
# Step 3 : Detect Outliers using IQR Method
# ----------------------------------------------------------

# Empty list to store results
outlier_results = []

# Loop through every feature
for feature in feature_columns:

    # Calculate first quartile (Q1)
    Q1 = train_df[feature].quantile(0.25)

    # Calculate third quartile (Q3)
    Q3 = train_df[feature].quantile(0.75)

    # Calculate IQR
    IQR = Q3 - Q1

    # Lower bound
    lower_bound = Q1 - 1.5 * IQR

    # Upper bound
    upper_bound = Q3 + 1.5 * IQR

    # Detect outliers
    outliers = train_df[
        (train_df[feature] < lower_bound) |
        (train_df[feature] > upper_bound)
    ]

    # Count outliers
    outlier_count = len(outliers)

    # Calculate percentage
    outlier_percentage = (
        outlier_count /
        len(train_df)
    ) * 100

    # Store results
    outlier_results.append({
        "Feature": feature,
        "Outlier Count": outlier_count,
        "Outlier Percentage": round(outlier_percentage, 2)
    })

# ----------------------------------------------------------
# Step 4 : Create Summary Table
# ----------------------------------------------------------

# Convert list into DataFrame
outlier_df = pd.DataFrame(outlier_results)

# Sort by highest number of outliers
outlier_df = outlier_df.sort_values(
    by="Outlier Count",
    ascending=False
).reset_index(drop=True)

# ----------------------------------------------------------
# Step 5 : Display Results
# ----------------------------------------------------------

print("=" * 75)
print("Outlier Summary using IQR Method")
print("=" * 75)

display(outlier_df)

# ----------------------------------------------------------
# Step 6 : Display Overall Summary
# ----------------------------------------------------------

print("\nTop 10 Features with Highest Number of Outliers")

display(outlier_df.head(10))
```

# Data Preprocessing - Train and Validation Split

## Objective

Before training machine learning models, the dataset must be divided into separate training and validation sets.

The **training set** is used to train the model, while the **validation set** is used to evaluate how well the model performs on unseen data.

---

## Why use a Stratified Split?

The ReneWind dataset is **highly imbalanced**, with generator failures representing only a small percentage of all observations.

If the data were split randomly, one dataset might contain significantly fewer failure cases than the other. This could lead to:

- Poor model training
- Unreliable evaluation metrics
- Biased performance estimates

To avoid this problem, we use **stratified sampling**, which preserves the original class distribution in both the training and validation datasets.

This ensures that both datasets contain approximately the same proportion of failure and non-failure observations.

---

## Split Configuration

- **Training Set:** 80%
- **Validation Set:** 20%
- **Sampling Method:** Stratified Sampling
- **Random State:** 42 (for reproducibility)

```python
# ==========================================================
# Data Preprocessing
# Train-Validation Split
# ==========================================================

# ----------------------------------------------------------
# Step 1 : Separate Features and Target
# ----------------------------------------------------------

# Create feature matrix (X) by removing the target column
X = train_df.drop(columns=["Target"])

# Create target vector (y)
y = train_df["Target"]

# ----------------------------------------------------------
# Step 2 : Perform Stratified Train-Validation Split
# ----------------------------------------------------------

# Split the dataset into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(
    X,
    y,
    test_size=0.20,          # 20% validation data
    stratify=y,              # Preserve class distribution
    random_state=42          # Ensure reproducibility
)

# ----------------------------------------------------------
# Step 3 : Print Dataset Shapes
# ----------------------------------------------------------

print("=" * 60)
print("Dataset Shapes")
print("=" * 60)

print(f"X_train : {X_train.shape}")
print(f"X_val   : {X_val.shape}")
print(f"y_train : {y_train.shape}")
print(f"y_val   : {y_val.shape}")

# ----------------------------------------------------------
# Step 4 : Display Class Distribution
# ----------------------------------------------------------

print("\n" + "=" * 60)
print("Class Distribution in Training Set")
print("=" * 60)

train_distribution = (
    y_train.value_counts()
    .rename_axis("Class")
    .reset_index(name="Count")
)

train_distribution["Percentage"] = (
    train_distribution["Count"] / len(y_train) * 100
).round(2)

display(train_distribution)

print("\n" + "=" * 60)
print("Class Distribution in Validation Set")
print("=" * 60)

val_distribution = (
    y_val.value_counts()
    .rename_axis("Class")
    .reset_index(name="Count")
)

val_distribution["Percentage"] = (
    val_distribution["Count"] / len(y_val) * 100
).round(2)

display(val_distribution)

# ----------------------------------------------------------
# Step 5 : Verify Stratification
# ----------------------------------------------------------

print("\n" + "=" * 60)
print("Verification")
print("=" * 60)

print(
    f"Original Failure Rate   : {y.mean() * 100:.2f}%"
)

print(
    f"Training Failure Rate   : {y_train.mean() * 100:.2f}%"
)

print(
    f"Validation Failure Rate : {y_val.mean() * 100:.2f}%"
)

print("\n✅ Stratified split completed successfully.")
```

# Data Preprocessing - Missing Value Handling

## Objective

Before training machine learning models, we must ensure that the dataset contains no missing values.

Many machine learning algorithms cannot handle missing values directly, making this an essential preprocessing step.

In this section, we will:

1. Check for missing values in the training, validation, and test datasets.
2. If missing values exist, replace them using **Median Imputation**.
3. If no missing values are found, keep the pipeline unchanged while documenting that no imputation was necessary.

---

## Why fit the imputer only on the training data?

The imputer learns the replacement values (medians) from the training data.

These learned values are then applied to:

- Validation data
- Test data

This prevents **data leakage**, where information from unseen data influences the training process.

Avoiding data leakage ensures that model evaluation remains realistic and accurately reflects real-world performance.

```python
# ==========================================================
# Data Preprocessing
# Missing Value Handling
# ==========================================================

# ----------------------------------------------------------
# Step 1 : Import SimpleImputer
# ----------------------------------------------------------

from sklearn.impute import SimpleImputer

# ----------------------------------------------------------
# Step 2 : Check Missing Values
# ----------------------------------------------------------

# Count missing values in each dataset
train_missing = X_train.isnull().sum().sum()
val_missing = X_val.isnull().sum().sum()
test_missing = test_df.isnull().sum().sum()

# Display results
print("=" * 65)
print("Missing Value Summary")
print("=" * 65)

print(f"Training Set Missing Values   : {train_missing}")
print(f"Validation Set Missing Values : {val_missing}")
print(f"Test Set Missing Values       : {test_missing}")

# ----------------------------------------------------------
# Step 3 : Apply Median Imputation (if required)
# ----------------------------------------------------------

# Check whether any dataset contains missing values
if (train_missing > 0) or (val_missing > 0) or (test_missing > 0):

    print("\nMissing values detected.")
    print("Applying Median Imputation...")

    # Create SimpleImputer object
    imputer = SimpleImputer(strategy="median")

    # Learn medians ONLY from training data
    imputer.fit(X_train)

    # Transform training data
    X_train = pd.DataFrame(
        imputer.transform(X_train),
        columns=X_train.columns,
        index=X_train.index
    )

    # Transform validation data
    X_val = pd.DataFrame(
        imputer.transform(X_val),
        columns=X_val.columns,
        index=X_val.index
    )

    # Make sure test data has only the feature columns
    test_features = test_df[X_train.columns]

    test_df = pd.DataFrame(
    imputer.transform(test_features),
    columns=X_train.columns,
    index=test_df.index
)

    print("\nMedian imputation completed successfully.")

# ----------------------------------------------------------
# Step 4 : No Missing Values Found
# ----------------------------------------------------------

else:

    print("\nNo missing values were detected.")
    print("No imputation is required.")

    print("\nPipeline Note:")
    print("- This preprocessing step is intentionally retained.")
    print("- Production machine learning pipelines should always")
    print("  include missing value handling to ensure robustness.")
    print("- Future datasets may contain missing values even if")
    print("  the current training dataset does not.")

# ----------------------------------------------------------
# Step 5 : Verify Missing Values After Processing
# ----------------------------------------------------------

print("\n" + "=" * 65)
print("Verification After Preprocessing")
print("=" * 65)

print(f"Training Set Missing Values   : {X_train.isnull().sum().sum()}")
print(f"Validation Set Missing Values : {X_val.isnull().sum().sum()}")
print(f"Test Set Missing Values       : {test_df.isnull().sum().sum()}")

print("\n✅ Missing value handling completed.")
```

# Interpretation

The missing value handling step ensures that all datasets are suitable for machine learning.

### If Missing Values Were Present

Median imputation was applied using statistics learned **only from the training data**.

Using the training data to compute replacement values prevents **data leakage**, ensuring that no information from the validation or test datasets influences model training.

Median imputation is preferred because it is robust to extreme values and preserves the overall distribution of the data better than mean imputation when outliers are present.

---

### If No Missing Values Were Present

No imputation was required because the dataset was already complete.

Nevertheless, this preprocessing step remains part of the machine learning pipeline because:

- Production datasets frequently contain missing values.
- Future data collected from sensors may include incomplete observations.
- Keeping the preprocessing pipeline consistent improves reproducibility and makes the workflow easier to deploy in real-world environments.

Including this step ensures that the pipeline is robust, reusable, and ready for production use.

# Data Preprocessing - Feature Scaling

## Objective

Machine learning algorithms often perform better when the input features are on a similar scale.

In this section, we standardize all numerical features using **StandardScaler**.

The scaler is:

- **Fitted only on the training data**
- Applied to the training, validation, and test datasets using the same learned parameters

---

## Why is Feature Scaling Important?

Different sensor features may have different ranges and units.

For example:

- One sensor may measure values between **0 and 1**.
- Another may measure values between **100 and 10,000**.

Without scaling, features with larger numerical values can dominate the learning process, causing slower or less effective training.

---

## Why is Scaling Essential for Neural Networks?

Neural Networks train using gradient-based optimization algorithms.

When features have very different scales:

- Training becomes slower.
- Gradient updates become unstable.
- Model convergence may be poor.

Scaling helps:

- Speed up convergence.
- Improve numerical stability.
- Produce more reliable model performance.

---

## Why Fit the Scaler Only on the Training Data?

The scaler computes the **mean** and **standard deviation** from the training dataset.

These same statistics are then applied to:

- Validation data
- Test data

This prevents **data leakage**, ensuring that information from unseen data does not influence model training.

---

## Why StandardScaler Instead of MinMaxScaler?

We use **StandardScaler** because it transforms each feature to have:

- Mean ≈ 0
- Standard Deviation ≈ 1

This method works well for many machine learning algorithms, particularly:

- Logistic Regression
- Support Vector Machines (SVM)
- Neural Networks

Unlike MinMaxScaler, StandardScaler is generally less sensitive to the presence of outliers, making it a suitable choice for sensor data where extreme values may represent genuine equipment behavior.

```python
# ==========================================================
# Data Preprocessing
# Feature Scaling using StandardScaler
# ==========================================================

# ----------------------------------------------------------
# Step 1 : Import StandardScaler
# ----------------------------------------------------------

from sklearn.preprocessing import StandardScaler

# ----------------------------------------------------------
# Step 2 : Display Statistics Before Scaling
# ----------------------------------------------------------

print("=" * 75)
print("Feature Statistics BEFORE Scaling (First 5 Features)")
print("=" * 75)

before_scaling = pd.DataFrame({
    "Mean": X_train.iloc[:, :5].mean(),
    "Std Dev": X_train.iloc[:, :5].std()
})

display(before_scaling)

# ----------------------------------------------------------
# Step 3 : Create StandardScaler Object
# ----------------------------------------------------------

# Initialize the scaler
scaler = StandardScaler()

# ----------------------------------------------------------
# Step 4 : Fit Scaler ONLY on Training Data
# ----------------------------------------------------------

# Learn mean and standard deviation from training data
scaler.fit(X_train)

# ----------------------------------------------------------
# Step 5 : Transform Datasets
# ----------------------------------------------------------

# Scale training features
X_train_scaled = scaler.transform(X_train)

# Scale validation features
X_val_scaled = scaler.transform(X_val)

# Scale test features
test_scaled = scaler.transform(test_df)

# ----------------------------------------------------------
# Step 6 : Convert Arrays Back to DataFrames
# ----------------------------------------------------------

# Preserve column names and indices
X_train_scaled = pd.DataFrame(
    X_train_scaled,
    columns=X_train.columns,
    index=X_train.index
)

X_val_scaled = pd.DataFrame(
    X_val_scaled,
    columns=X_val.columns,
    index=X_val.index
)

test_scaled = pd.DataFrame(
    test_scaled,
    columns=test_df.columns,
    index=test_df.index
)

# ----------------------------------------------------------
# Step 7 : Display Statistics After Scaling
# ----------------------------------------------------------

print("\n" + "=" * 75)
print("Feature Statistics AFTER Scaling (First 5 Features)")
print("=" * 75)

after_scaling = pd.DataFrame({
    "Mean": X_train_scaled.iloc[:, :5].mean(),
    "Std Dev": X_train_scaled.iloc[:, :5].std()
})

display(after_scaling)

# ----------------------------------------------------------
# Step 8 : Verification
# ----------------------------------------------------------

print("\n" + "=" * 75)
print("Verification")
print("=" * 75)

print("Training Shape   :", X_train_scaled.shape)
print("Validation Shape :", X_val_scaled.shape)
print("Test Shape       :", test_scaled.shape)

print("\n✅ Feature scaling completed successfully.")
print("The original datasets (X_train, X_val, test_df) remain unchanged.")
print("The scaled datasets (X_train_scaled, X_val_scaled, test_scaled) are ready for models that require standardized inputs.")
```

# Data Preprocessing - Handling Class Imbalance

## Objective

The ReneWind dataset is highly imbalanced, with generator failures representing only a small proportion of the training data.

If this imbalance is ignored, a machine learning model may become biased toward predicting the majority class (No Failure), resulting in poor detection of actual failures.

To address this problem, we will prepare two different imbalance handling strategies that can later be compared during model development.

---

## Approach 1 — Class Weights

Class weighting does **not** modify the training data.

Instead, it assigns a larger penalty to mistakes made on the minority class.

During training:

- Errors on the **Failure** class receive a higher penalty.
- Errors on the **No Failure** class receive a lower penalty.

This encourages the model to pay greater attention to the minority class without creating synthetic observations.

---

## Approach 2 — SMOTE (Synthetic Minority Oversampling Technique)

SMOTE increases the number of minority-class observations by generating **synthetic samples**.

Rather than duplicating existing observations, SMOTE creates new examples by interpolating between nearby minority-class samples.

The resulting training dataset becomes balanced while preserving much of the original feature space.

---

## Why are these methods applied only to the training data?

The validation and test datasets must represent real-world data.

Applying class weighting or SMOTE to validation or test data would produce unrealistic evaluation results and introduce **data leakage**.

Only the training data should be modified or reweighted.

---

## Which approach will be our primary method?

For the ReneWind project, **Class Weights** will be the primary imbalance handling technique because:

- It preserves the original data distribution.
- It avoids creating synthetic sensor readings.
- It works well for many machine learning algorithms.
- It is particularly appropriate for predictive maintenance datasets where unusual sensor values may represent genuine equipment behaviour.

SMOTE will also be prepared so that its performance can later be compared against the class-weight approach.

---

## When is SMOTE preferred?

SMOTE is often beneficial when:

- The minority class is extremely small.
- The model struggles to learn minority-class patterns.
- Additional minority examples improve decision boundaries.

However, because SMOTE generates artificial observations, it should always be applied **after** the train/validation split and **only** to the training data.

```python
# ==========================================================
# Data Preprocessing
# Handling Class Imbalance
# ==========================================================

# ----------------------------------------------------------
# Step 1 : Import Required Libraries
# ----------------------------------------------------------

from sklearn.utils.class_weight import compute_class_weight
from imblearn.over_sampling import SMOTE
import numpy as np

# ----------------------------------------------------------
# Approach 1 : Compute Class Weights
# ----------------------------------------------------------

print("=" * 70)
print("Approach 1 : Class Weights")
print("=" * 70)

# Compute class weights
weights = compute_class_weight(
    class_weight="balanced",
    classes=np.unique(y_train),
    y=y_train
)

# Create dictionary
class_weight = {
    0: weights[0],
    1: weights[1]
}

# Display class weights
print("Computed Class Weights\n")
print(class_weight)

# ----------------------------------------------------------
# Approach 2 : SMOTE
# ----------------------------------------------------------

print("\n" + "=" * 70)
print("Approach 2 : SMOTE")
print("=" * 70)

# Display class distribution BEFORE SMOTE
print("\nClass Distribution Before SMOTE\n")

before_smote = (
    y_train.value_counts()
    .rename_axis("Class")
    .reset_index(name="Count")
)

before_smote["Percentage"] = (
    before_smote["Count"] / len(y_train) * 100
).round(2)

display(before_smote)

# ----------------------------------------------------------
# Apply SMOTE ONLY on Training Data
# ----------------------------------------------------------

# Create SMOTE object
smote = SMOTE(
    random_state=42
)

# Generate synthetic samples
X_train_smote, y_train_smote = smote.fit_resample(
    X_train,
    y_train
)

# ----------------------------------------------------------
# Display Class Distribution AFTER SMOTE
# ----------------------------------------------------------

print("Class Distribution After SMOTE\n")

after_smote = (
    pd.Series(y_train_smote)
    .value_counts()
    .rename_axis("Class")
    .reset_index(name="Count")
)

after_smote["Percentage"] = (
    after_smote["Count"] /
    len(y_train_smote) * 100
).round(2)

display(after_smote)

# ----------------------------------------------------------
# Display Dataset Shapes
# ----------------------------------------------------------

print("=" * 70)
print("Dataset Shapes")
print("=" * 70)

print("Original Training Set :", X_train.shape)
print("SMOTE Training Set    :", X_train_smote.shape)

print("\n✅ Class imbalance handling completed.")
```

# Interpretation

Two approaches have been prepared to address the class imbalance in the ReneWind dataset.

## Approach 1 — Class Weights

Class weighting increases the importance of the minority class during model training without modifying the original data.

This approach:

- Preserves the original distribution of sensor readings.
- Avoids generating artificial observations.
- Is widely used in predictive maintenance and industrial machine learning applications.

For this project, **Class Weights will be the primary imbalance handling strategy**.

## Approach 2 — SMOTE

SMOTE balances the training dataset by generating synthetic failure examples.

After applying SMOTE:

Only the **training dataset** was oversampled.

The validation and test datasets remain unchanged to ensure that model evaluation reflects real-world operating conditions.

# Model Evaluation Framework

## Objective

Before training machine learning models, it is important to define a consistent evaluation framework.

Every model in this project will be evaluated using the same metrics so that their performance can be compared fairly.

---

## Why is Recall Our Primary Metric?

The ReneWind dataset is highly imbalanced, and the business cost of missing a generator failure is extremely high.

Business Costs:

- False Negative (FN) = **\$200,000 Loss**
- False Positive (FP) = **\$5,000 Cost**
- True Positive (TP) = **\$20,000 Benefit**

A False Negative means that the model predicts **"No Failure"** when the turbine is actually going to fail.

Since missing a failure is far more expensive than generating a false alarm, maximizing **Recall** for the failure class is the primary objective.

---

## Why Accuracy is Misleading

Suppose only 5% of turbines fail.

A model that predicts **"No Failure"** for every turbine would achieve approximately **95% accuracy**, yet it would detect **zero actual failures**.

Although the accuracy appears high, such a model would be practically useless because it completely fails to identify generator failures.

For this reason, accuracy will be reported but will not be used as the primary evaluation metric.

---

## Business Cost

To evaluate models from a business perspective, we compute:

Business Cost

= (False Negatives × \$200,000)
+ (False Positives × \$5,000)
− (True Positives × \$20,000)

Lower business cost indicates a better predictive maintenance model because it captures more failures while minimizing unnecessary maintenance actions.

```python
# ==========================================================
# Model Evaluation Framework
# ==========================================================

# ----------------------------------------------------------
# Step 1 : Import Required Libraries
# ----------------------------------------------------------

from sklearn.metrics import (
    confusion_matrix,
    classification_report,
    roc_auc_score,
    average_precision_score,
    ConfusionMatrixDisplay
)

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# ----------------------------------------------------------
# Step 2 : Evaluation Function
# ----------------------------------------------------------

def evaluate_model(y_true, y_prob, threshold=0.50):

    """
    Evaluate a binary classification model.

    Parameters
    ----------
    y_true : Actual labels

    y_prob : Predicted probabilities

    threshold : Decision threshold

    Returns
    -------
    Dictionary containing evaluation metrics
    """

    # --------------------------------------------
    # Convert probabilities into predictions
    # --------------------------------------------

    y_pred = (y_prob >= threshold).astype(int)

    # --------------------------------------------
    # Confusion Matrix
    # --------------------------------------------

    cm = confusion_matrix(y_true, y_pred)

    TN, FP, FN, TP = cm.ravel()

    # --------------------------------------------
    # Plot Confusion Matrix
    # --------------------------------------------

    plt.figure(figsize=(6,5))

    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="Blues",
        xticklabels=["No Failure","Failure"],
        yticklabels=["No Failure","Failure"]
    )

    plt.xlabel("Predicted")

    plt.ylabel("Actual")

    plt.title("Confusion Matrix")

    plt.show()

    # --------------------------------------------
    # Classification Report
    # --------------------------------------------

    print("="*70)
    print("Classification Report")
    print("="*70)

    print(classification_report(y_true,y_pred))

    # --------------------------------------------
    # ROC-AUC
    # --------------------------------------------

    roc_auc = roc_auc_score(
        y_true,
        y_prob
    )

    # --------------------------------------------
    # Precision-Recall AUC
    # --------------------------------------------

    pr_auc = average_precision_score(
        y_true,
        y_prob
    )

    # --------------------------------------------
    # Business Cost
    # --------------------------------------------

    business_cost = (

        FN * 200000

        +

        FP * 5000

        -

        TP * 20000

    )

    # --------------------------------------------
    # Display Metrics
    # --------------------------------------------

    print("="*70)

    print(f"ROC-AUC Score          : {roc_auc:.4f}")

    print(f"PR-AUC Score           : {pr_auc:.4f}")

    print(f"Business Cost ($)      : {business_cost:,.0f}")

    print("="*70)

    # --------------------------------------------
    # Return Metrics
    # --------------------------------------------

    return {

        "TN":TN,

        "FP":FP,

        "FN":FN,

        "TP":TP,

        "ROC_AUC":roc_auc,

        "PR_AUC":pr_auc,

        "Business_Cost":business_cost

    }
```

# Neural Network - Architecture 1

## Objective

The first deep learning model serves as the baseline neural network for the ReneWind predictive maintenance project.

The objective is to learn the relationship between the 40 sensor measurements and the probability of generator failure.

---

## Network Architecture

The model consists of:

- **Input Layer:** 40 numerical sensor features
- **Hidden Layer 1:** 64 neurons with ReLU activation
- **Dropout Layer:** 20%
- **Hidden Layer 2:** 32 neurons with ReLU activation
- **Dropout Layer:** 20%
- **Output Layer:** 1 neuron with Sigmoid activation

---

## Why ReLU?

The Rectified Linear Unit (ReLU) is widely used because it:

- Learns complex non-linear relationships
- Trains efficiently
- Reduces the vanishing gradient problem

---

## Why Dropout?

Dropout randomly deactivates a percentage of neurons during training.

Benefits include:

- Reducing overfitting
- Improving generalization
- Making the network more robust

A dropout rate of **20%** is used after each hidden layer.

---

## Why Sigmoid?

Since this is a **binary classification** problem, the output layer contains a single neuron with a **sigmoid activation**, producing a probability between **0 and 1** that represents the likelihood of generator failure.

---

## Training Configuration

The model is trained using:

- Optimizer: Adam
- Learning Rate: 0.001
- Loss Function: Binary Crossentropy
- Metrics:
  - Recall
  - Precision
  - AUC

To address the class imbalance, the **class weights** computed during preprocessing are supplied during training.

Early stopping monitors **Validation Recall**, allowing the model to stop training once performance stops improving while restoring the best-performing weights.

```python
# ==========================================================
# Architecture 1
# ==========================================================

# ----------------------------------------------------------
# Import Required Libraries
# ----------------------------------------------------------

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
import tensorflow as tf

# ----------------------------------------------------------
# Build Neural Network
# ----------------------------------------------------------

model_arch1 = Sequential()

# Input Layer + Hidden Layer 1
model_arch1.add(
    Dense(
        64,
        activation="relu",
        input_shape=(40,)
    )
)

# Dropout Layer
model_arch1.add(
    Dropout(0.20)
)

# Hidden Layer 2
model_arch1.add(
    Dense(
        32,
        activation="relu"
    )
)

# Dropout Layer
model_arch1.add(
    Dropout(0.20)
)

# Output Layer
model_arch1.add(
    Dense(
        1,
        activation="sigmoid"
    )
)

# ----------------------------------------------------------
# Compile Model
# ----------------------------------------------------------

model_arch1.compile(

    optimizer=Adam(
        learning_rate=0.001
    ),

    loss="binary_crossentropy",

    metrics=[
        tf.keras.metrics.Recall(name="recall"),
        tf.keras.metrics.Precision(name="precision"),
        tf.keras.metrics.AUC(name="auc")
    ]

)

# ----------------------------------------------------------
# Display Model Summary
# ----------------------------------------------------------

print("="*70)
print("Architecture 1 Summary")
print("="*70)

model_arch1.summary()

# ----------------------------------------------------------
# Early Stopping
# ----------------------------------------------------------

early_stopping = EarlyStopping(

    monitor="val_recall",

    mode="max",

    patience=10,

    restore_best_weights=True

)

# ----------------------------------------------------------
# Train Model
# ----------------------------------------------------------

history_arch1 = model_arch1.fit(

    X_train_scaled,

    y_train,

    validation_data=(

        X_val_scaled,

        y_val

    ),

    epochs=100,

    batch_size=32,

    class_weight=class_weight,

    callbacks=[early_stopping],

    verbose=1

)

# ----------------------------------------------------------
# Plot Training History
# ----------------------------------------------------------

plot_training_history(history_arch1)
```

# Neural Network - Architecture 2 (Primary Candidate)

## Objective

Architecture 2 improves upon the baseline neural network by introducing **Batch Normalization** and a slightly deeper architecture.

The goal is to improve training stability, reduce overfitting, and achieve better generalization on unseen turbine data.

---

## Network Architecture

The model consists of:

- **Input Layer:** 40 numerical sensor features
- **Hidden Layer 1:** 64 neurons (ReLU)
- **Batch Normalization**
- **Dropout:** 30%
- **Hidden Layer 2:** 32 neurons (ReLU)
- **Batch Normalization**
- **Dropout:** 30%
- **Hidden Layer 3:** 16 neurons (ReLU)
- **Dropout:** 20%
- **Output Layer:** 1 neuron (Sigmoid)

---

## What does Batch Normalization add?

Architecture 1 relied only on Dropout for regularization.

Architecture 2 additionally uses **Batch Normalization**, which:

- Normalizes activations during training
- Stabilizes gradient updates
- Allows faster convergence
- Makes training less sensitive to weight initialization
- Often improves validation performance

---

## Why use ReduceLROnPlateau?

As training progresses, improvements often become smaller.

ReduceLROnPlateau automatically decreases the learning rate when validation loss stops improving.

This enables the optimizer to make smaller and more precise updates, often leading to better convergence.

---

## Why is this our Primary Candidate?

Compared to Architecture 1, this model includes:

- Batch Normalization for improved stability
- A deeper feature extraction pipeline
- Stronger regularization through Dropout
- Adaptive learning rate scheduling

These improvements make Architecture 2 a strong candidate for achieving better predictive performance while maintaining good generalization.

```python
# ==========================================================
# Architecture 2
# ==========================================================

# ----------------------------------------------------------
# Import Required Libraries
# ----------------------------------------------------------

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import (
    EarlyStopping,
    ReduceLROnPlateau
)

import tensorflow as tf

# ----------------------------------------------------------
# Build Neural Network
# ----------------------------------------------------------

model_arch2 = Sequential()

# -----------------------------
# Input + Hidden Layer 1
# -----------------------------
model_arch2.add(
    Dense(
        64,
        activation="relu",
        input_shape=(40,)
    )
)

model_arch2.add(
    BatchNormalization()
)

model_arch2.add(
    Dropout(0.30)
)

# -----------------------------
# Hidden Layer 2
# -----------------------------
model_arch2.add(
    Dense(
        32,
        activation="relu"
    )
)

model_arch2.add(
    BatchNormalization()
)

model_arch2.add(
    Dropout(0.30)
)

# -----------------------------
# Hidden Layer 3
# -----------------------------
model_arch2.add(
    Dense(
        16,
        activation="relu"
    )
)

model_arch2.add(
    Dropout(0.20)
)

# -----------------------------
# Output Layer
# -----------------------------
model_arch2.add(
    Dense(
        1,
        activation="sigmoid"
    )
)

# ----------------------------------------------------------
# Compile Model
# ----------------------------------------------------------

model_arch2.compile(

    optimizer=Adam(
        learning_rate=0.001
    ),

    loss="binary_crossentropy",

    metrics=[

        tf.keras.metrics.Recall(
            name="recall"
        ),

        tf.keras.metrics.Precision(
            name="precision"
        ),

        tf.keras.metrics.AUC(
            name="auc"
        )

    ]

)

# ----------------------------------------------------------
# Display Model Summary
# ----------------------------------------------------------

print("=" * 70)
print("Architecture 2 Summary")
print("=" * 70)

model_arch2.summary()

# ----------------------------------------------------------
# Callbacks
# ----------------------------------------------------------

early_stopping = EarlyStopping(

    monitor="val_recall",

    mode="max",

    patience=10,

    restore_best_weights=True

)

reduce_lr = ReduceLROnPlateau(

    monitor="val_loss",

    factor=0.5,

    patience=5,

    verbose=1

)

# ----------------------------------------------------------
# Train Model
# ----------------------------------------------------------

history_arch2 = model_arch2.fit(

    X_train_scaled,

    y_train,

    validation_data=(

        X_val_scaled,

        y_val

    ),

    epochs=100,

    batch_size=32,

    class_weight=class_weight,

    callbacks=[

        early_stopping,

        reduce_lr

    ],

    verbose=1

)

# ----------------------------------------------------------
# Plot Training History
# ----------------------------------------------------------

plot_training_history(history_arch2)
```

# Neural Network - Architecture 3 (Advanced Regularized Model)

## Objective

Architecture 3 is the most advanced neural network developed for the ReneWind project.

It introduces **L2 Regularization** together with **Stochastic Gradient Descent (SGD) with Momentum** to investigate whether stronger regularization and a different optimization strategy improve the model's ability to generalize to unseen turbine data.

This architecture is designed to reduce overfitting while maintaining high recall for generator failure detection.

---

## Network Architecture

The model consists of:

- **Input Layer:** 40 numerical sensor features
- **Hidden Layer 1:** 64 neurons (ReLU + L2 Regularization)
- **Batch Normalization**
- **Dropout:** 30%
- **Hidden Layer 2:** 64 neurons (ReLU + L2 Regularization)
- **Dropout:** 30%
- **Hidden Layer 3:** 32 neurons (ReLU + L2 Regularization)
- **Dropout:** 20%
- **Hidden Layer 4:** 16 neurons (ReLU)
- **Output Layer:** 1 neuron (Sigmoid)

---

## What does L2 Regularization do?

L2 Regularization adds a penalty to the loss function based on the magnitude of the model's weights.

Instead of allowing very large weights, L2 encourages the network to learn smaller and more balanced weights.

Benefits include:

- Reduces overfitting
- Improves model generalization
- Produces smoother decision boundaries
- Helps prevent overly complex models

---

## Why use SGD with Momentum?

The previous architectures used the Adam optimizer.

In this architecture, we evaluate **Stochastic Gradient Descent (SGD) with Momentum**.

Momentum helps accelerate learning by carrying information from previous gradient updates.

Benefits include:

- Smoother optimization
- Reduced oscillations during training
- Better convergence on some deep learning problems
- Opportunity to compare two commonly used optimization algorithms

Comparing Adam and SGD allows us to determine which optimizer performs better for the ReneWind predictive maintenance problem.

```python
# ==========================================================
# Architecture 3
# ==========================================================

# ----------------------------------------------------------
# Import Required Libraries
# ----------------------------------------------------------

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.regularizers import l2
import tensorflow as tf

# ----------------------------------------------------------
# Build Neural Network
# ----------------------------------------------------------

model_arch3 = Sequential()

# Hidden Layer 1
model_arch3.add(
    Dense(
        64,
        activation="relu",
        kernel_regularizer=l2(0.001),
        input_shape=(40,)
    )
)

model_arch3.add(BatchNormalization())
model_arch3.add(Dropout(0.30))

# Hidden Layer 2
model_arch3.add(
    Dense(
        64,
        activation="relu",
        kernel_regularizer=l2(0.001)
    )
)

model_arch3.add(Dropout(0.30))

# Hidden Layer 3
model_arch3.add(
    Dense(
        32,
        activation="relu",
        kernel_regularizer=l2(0.001)
    )
)

model_arch3.add(Dropout(0.20))

# Hidden Layer 4
model_arch3.add(
    Dense(
        16,
        activation="relu"
    )
)

# Output Layer
model_arch3.add(
    Dense(
        1,
        activation="sigmoid"
    )
)

# ----------------------------------------------------------
# Compile Model
# ----------------------------------------------------------

model_arch3.compile(

    optimizer=SGD(
        learning_rate=0.01,
        momentum=0.9
    ),

    loss="binary_crossentropy",

    metrics=[
        tf.keras.metrics.Recall(name="recall"),
        tf.keras.metrics.Precision(name="precision"),
        tf.keras.metrics.AUC(name="auc")
    ]
)

# ----------------------------------------------------------
# Display Model Summary
# ----------------------------------------------------------

print("="*70)
print("Architecture 3 Summary")
print("="*70)

model_arch3.summary()

# ----------------------------------------------------------
# Early Stopping
# ----------------------------------------------------------

early_stopping = EarlyStopping(

    monitor="val_recall",

    mode="max",

    patience=10,

    restore_best_weights=True

)

# ----------------------------------------------------------
# Train Model
# ----------------------------------------------------------

history_arch3 = model_arch3.fit(

    X_train_scaled,

    y_train,

    validation_data=(

        X_val_scaled,

        y_val

    ),

    epochs=100,

    batch_size=32,

    class_weight=class_weight,

    callbacks=[early_stopping],

    verbose=1

)

# ----------------------------------------------------------
# Plot Training History
# ----------------------------------------------------------

plot_training_history(history_arch3)
```
