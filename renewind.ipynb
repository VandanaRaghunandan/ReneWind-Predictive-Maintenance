# Exploratory Data Analysis (EDA) - Class Distribution

## Objective
Before building any machine learning model, it is important to understand how the target classes are distributed.

The target variable represents:

- **0** → No Generator Failure
- **1** → Generator Failure

In real-world predictive maintenance datasets, failure events are usually much rarer than normal operating conditions. This results in an **imbalanced dataset**, where one class significantly outnumbers the other.

### Why is Class Imbalance a Problem?

For the ReneWind project, class imbalance is particularly important because the business costs are highly unequal.

| Prediction Outcome | Business Cost |
|--------------------|--------------:|
| False Negative (Missed Failure) | **$200,000 Loss** |
| False Positive (False Alarm) | **$5,000 Cost** |
| True Positive (Correct Failure Prediction) | **$20,000 Benefit** |

A model that simply predicts every turbine as **"No Failure"** may achieve a high accuracy because failures are rare, but it would miss most actual failures. Each missed failure can lead to catastrophic financial losses.

Therefore, throughout this project, we will focus on identifying failures effectively rather than relying solely on overall accuracy.

```python
# ==========================================================
# Exploratory Data Analysis (EDA)
# Class Distribution Analysis
# ==========================================================

# ----------------------------------------------------------
# Step 1 : Calculate Class Distribution (Raw Counts)
# ----------------------------------------------------------

# Count the number of observations in each target class
class_counts = train_df['Target'].value_counts()

# Display raw counts
print("=" * 50)
print("Class Distribution (Count)")
print("=" * 50)
print(class_counts)

# ----------------------------------------------------------
# Step 2 : Calculate Class Distribution (Percentage)
# ----------------------------------------------------------

# Convert counts into percentages
class_percentages = train_df['Target'].value_counts(normalize=True) * 100

# Display percentages
print("\n" + "=" * 50)
print("Class Distribution (Percentage)")
print("=" * 50)
print(class_percentages.round(2))

# ----------------------------------------------------------
# Step 3 : Visualize Class Distribution
# ----------------------------------------------------------

# Create figure
plt.figure(figsize=(7,5))

# Draw countplot
ax = sns.countplot(
    x='Target',
    data=train_df,
    palette='Set2'
)

# Set title
plt.title("Class Distribution of Generator Failure", fontsize=14)

# Set axis labels
plt.xlabel("Target Class")
plt.ylabel("Number of Samples")

# Rename x-axis labels
ax.set_xticklabels(["No Failure (0)", "Failure (1)"])

# ----------------------------------------------------------
# Step 4 : Display Exact Count on Top of Each Bar
# ----------------------------------------------------------

# Loop through each bar
for bar in ax.patches:

    # Get bar height
    height = bar.get_height()

    # Display value above bar
    ax.annotate(
        f'{int(height)}',
        (bar.get_x() + bar.get_width()/2, height),
        ha='center',
        va='bottom',
        fontsize=11,
        fontweight='bold'
    )

# Improve layout
plt.tight_layout()

# Display plot
plt.show()
```

## Interpretation

From the class distribution analysis, we observe that the dataset is **highly imbalanced**.

- The majority of observations belong to the **No Failure (0)** class.
- Only a small percentage of the data represents **Generator Failure (1)**.

This imbalance means that a machine learning model may become biased toward predicting the majority class, resulting in high overall accuracy but poor detection of actual failures.

For the ReneWind project, this is a critical issue because:

- Missing an actual failure (**False Negative**) results in a **$200,000 loss**.
- Raising a false alarm (**False Positive**) costs only **$5,000**.

Since the cost of missing a failure is much greater than the cost of a false alarm, our objective will be to build models that maximize the detection of failure cases while maintaining a reasonable false alarm rate.

In the next stages, we will evaluate suitable techniques such as **class weighting**, **SMOTE**, and **threshold tuning** to improve the model's ability to identify generator failures.

# Exploratory Data Analysis (EDA) - Data Quality Assessment

## Objective

Before building a machine learning model, it is important to evaluate the overall quality of the dataset.

In this section, we will:

1. Check for missing values in every feature.
2. Calculate the percentage of missing values.
3. Study the statistical summary of each feature.
4. Visualize missing values using a heatmap.

### Why is this important?

Machine learning models perform best when trained on clean, high-quality data.

Missing values can:
- Reduce model performance.
- Introduce bias into predictions.
- Cause some algorithms to fail during training.

The statistical summary helps us understand:
- Average values
- Data spread
- Minimum and maximum values
- Possible outliers
- Feature distributions

Understanding the dataset at this stage helps us decide whether preprocessing is required before model training.

```python
# ==========================================================
# Exploratory Data Analysis (EDA)
# Data Quality Assessment
# ==========================================================

# ----------------------------------------------------------
# Step 1 : Check Missing Values
# ----------------------------------------------------------

# Count the number of missing values in each column
missing_values = train_df.isnull().sum()

# Display missing values
print("=" * 60)
print("Missing Values Per Column")
print("=" * 60)
print(missing_values)

# ----------------------------------------------------------
# Step 2 : Calculate Percentage of Missing Values
# ----------------------------------------------------------

# Calculate percentage of missing values
missing_percentage = (train_df.isnull().sum() / len(train_df)) * 100

# Display percentage of missing values
print("\n" + "=" * 60)
print("Percentage of Missing Values")
print("=" * 60)
print(missing_percentage.round(2))

# ----------------------------------------------------------
# Step 3 : Check for Duplicate Records
# ----------------------------------------------------------

# Count duplicate rows
duplicate_rows = train_df.duplicated().sum()

# Display duplicate count
print("\n" + "=" * 60)
print("Duplicate Records")
print("=" * 60)
print(f"Number of Duplicate Rows : {duplicate_rows}")

# ----------------------------------------------------------
# Step 4 : Display Statistical Summary
# ----------------------------------------------------------

print("\n" + "=" * 60)
print("Statistical Summary of Numerical Features")
print("=" * 60)

# Display descriptive statistics
display(train_df.describe())

# ----------------------------------------------------------
# Step 5 : Visualize Missing Values
# ----------------------------------------------------------

# Create figure
plt.figure(figsize=(14,6))

# Plot heatmap of missing values
sns.heatmap(
    train_df.isnull(),
    cmap="viridis",
    cbar=True,
    yticklabels=False
)

# Add title
plt.title("Heatmap of Missing Values", fontsize=14)

# Label axes
plt.xlabel("Features")
plt.ylabel("Observations")

# Display heatmap
plt.show()

# ----------------------------------------------------------
# Step 6 : Overall Data Quality Summary
# ----------------------------------------------------------

# Calculate total missing values
total_missing = train_df.isnull().sum().sum()

print("\n" + "=" * 60)
print("Overall Data Quality Summary")
print("=" * 60)

print(f"Total Missing Values : {total_missing}")
print(f"Duplicate Records    : {duplicate_rows}")

# Display dataset status
if total_missing == 0 and duplicate_rows == 0:
    print("\n✅ The dataset is clean.")
    print("No missing values and no duplicate records were found.")
else:
    print("\n⚠️ The dataset requires preprocessing before model training.")
```

# Interpretation

### Missing Values

The missing value analysis helps determine whether any feature contains incomplete information.

- If all columns have **0 missing values**, no missing value treatment is required.
- If missing values exist, techniques such as mean, median, mode imputation, or advanced imputation methods may be considered depending on the feature and the percentage of missing data.

### Duplicate Records

Duplicate observations can negatively affect machine learning models by giving repeated samples more influence during training.

- If **0 duplicate rows** are found, no duplicate removal is required.
- If duplicates exist, they should generally be removed unless there is a valid business reason to keep them.

### Statistical Summary

The `describe()` output summarizes each numerical feature by providing:

- Count
- Mean
- Standard Deviation
- Minimum
- Quartiles (25%, 50%, 75%)
- Maximum

These statistics help identify:

- Features with large variation
- Possible outliers
- Features that may require scaling
- Unexpected value ranges

### Missing Value Heatmap

The heatmap provides a visual representation of missing data.

- A completely uniform heatmap indicates that the dataset contains no missing values.
- Colored gaps would indicate missing observations that require preprocessing.

### Overall Conclusion

If the dataset contains **no missing values** and **no duplicate records**, it is considered to have good data quality and is ready for the next stage of Exploratory Data Analysis (EDA), where we will study feature distributions, detect outliers, and analyze relationships between variables before building machine learning models.
